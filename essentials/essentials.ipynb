```python
images_dataset= tf.data.Dataset.list_files("training/*/*" , shuffle=False)## we are creating a dataset of image file paths
for file in images_dataset.take(3):## we are taking the first 5 file paths from the dataset and printing them
    print(file.numpy())## we are printing the file path as a numpy array
```

```python
training_size=int(image_count*0.7)# we are calculating the size of the training dataset by taking 70% of the total dataset size
train_ds=images_dataset.take(training_size)# we are taking the first 'training_size' file paths from the dataset for training
test_ds=images_dataset.skip(training_size)# we are skipping the first 'training_size' file paths and taking the rest for testing
```

```python
## we can create a funtşon to deddicate the class of the dataset
def get_label(file_path):
    import os
    return tf.strings.split(file_path,os.path.sep)[-2]
```

```python
def process_image(file_path):
    label=get_label(file_path)
    image=tf.io.read_file(file_path)
    image=tf.image.decode_jpeg(image, channels=3)
    image=tf.image.resize(image, [224,224])
    return image, label
```

```python
for label in train_ds.map(get_label).take(5):## we are applying the get_label function to the training dataset and taking the first 5 labels
    print(label.numpy())## we are printing the label value as a numpy array
```

```python
train_ds = train_ds.map(process_image)## we are applying the process_image function to the training dataset and returning a new dataset
for img,label in train_ds.take(3):## we are applying the process_image function to the training dataset and taking the first 5 images and labels
    print("image: ",img)
    print("label: ",label)
```

```python
def scale (image , label):
    return image/255.0, label
```

```python
train_ds=train_ds.map(scale)
for img,label in train_ds.take(3):
    print("image: ",img.numpy()[0][0])## we are printing the first pixel value of the first image as a numpy array
    print("label: ",label.numpy())
```
